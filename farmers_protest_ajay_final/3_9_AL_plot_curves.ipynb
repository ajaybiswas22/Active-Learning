```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

df = pd.read_csv('results/mlp_n2.csv')
df['Seed Sizes'] = df['Seed Sizes'].div(100).round(2)

df[['Seed Sizes', 'Accuracy', 'Precision', 'Recall',
    'F_Score']].multiply(100).plot(x='Seed Sizes')
plt.legend(['Accuracy', 'Precision', 'Recall', 'F_Score'])
plt.ylabel('Percentage')
plt.savefig('outputs/figures/' + 'MLP_20_5_666',dpi=300, bbox_inches='tight')
plt.show()
```
