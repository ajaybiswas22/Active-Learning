```python
from resources.basicIO import InputOutput as IO
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
import re
from itertools import islice, tee
import matplotlib.pyplot as plt
```

```python
df_sample = pd.read_csv('datasets/random_sample_data.csv')
# remove unnecessary comments
df_sample = df_sample.drop([x for x in range(len(df_sample)) if int(
    df_sample['label'][x]) != 0 and int(df_sample['label'][x]) != 1])

comments = df_sample['tokenized_comment']
labels = df_sample['label']
```

```python
uni = ['good','win','contract','election','land','kisan','middleman','money','corporates',
        'demand','consumer','family','sikh','modiji','suicide','victory','khalistani','road',
        'pakistan','monopoly','care','corporation','supporter','flag','khalistan','ambani',
        'adani']
bi = ['modi ji','i support','we support','jai kisan','private player',
      'rich farmer','free market','contract farm','private company','salute to','so call',
      'so called','big company','farmer want','700 farmer','only 6',
      'supreme court','these farmer','stand with','other state']
tri = ['punjab and haryana','for the farmer','do not want','farmer from punjab','good for farmer',
       'wheat and rice']
phrase_list = uni + bi + tri
coms = list(comments)
labs = list(labels)

for phrase in phrase_list:
    neg = 0
    pos = 0
    for i in range(len(coms)):
        if phrase in coms[i]:
            if(labs[i] == 0):
                neg += 1
            else:
                pos += 1
    print(phrase)
    print('Negative: ', neg)
    print('Positive: ', pos)
    print()
    
```
