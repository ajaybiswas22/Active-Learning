```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv("multipleComments.csv")
```

```python
display(df)
```

```python
def detect_language(line):
    maxchar = max(line)
    if u'\u0c00' <= maxchar <= u'\u0c7f':
        return 'telugu'
    elif u'\u0900' <= maxchar <= u'\u097f':
        return 'hindi'
    return 'english'
```

```python
# remove hindi rows
rows = []
eng_raw_df = df
for i in range(0,len(eng_raw_df)):
    line = eng_raw_df.Comment[i]
    if(detect_language(line) != 'hindi'):
        rows.append(i)

eng_raw_df = eng_raw_df.iloc[rows]
eng_raw_df.reset_index(inplace=True)
eng_raw_df = eng_raw_df.drop('index',1)
```

```python
display(eng_raw_df)
```

```python
# remove non ascii characters
eng_ascii_df = eng_raw_df
for i in range(0,len(eng_ascii_df)):
    line = eng_ascii_df.Comment[i]
    ascii_string = line.encode("ascii", "ignore")
    eng_ascii_df.at[i,'Comment'] = ascii_string.decode()
```

```python
eng_ascii_df.to_csv('englishComments.csv', index=False)
display(eng_ascii_df)
```
