```python
def getVideos(query, pages=None):
    video_list = []
    if pages is None:
        print('Setting default page limit to be 5')
        pages = 5
    else:
        print('Current query being processed is : ' + query)
        print('Page limit is set to ' + pages)
    npt = None
    for _ in range(pages):
        if npt is None:
            request = youtube.search().list(
                part = "snippet", 
                maxResults = 50,
                q = query,
                order="viewCount",
                videoDuration="any",
                videoType="any",
                safeSearch="none",
                relevanceLanguage='EN',
                regionCode='IN',
            )
        else:
            request = youtube.search().list(
                part = "snippet", 
                maxResults = 50,
                q = query,
                order="viewCount",
                videoDuration="any",
                videoType="any",
                safeSearch="none",
                relevanceLanguage='EN',
                regionCode='IN',
                pageToken=npt
            )
        response = request.execute()
        npt = response['nextPageToken']
        for x in range(len(response['items'])):
            video_list.append((response['items'][x]['id']['videoId'], response['items'][x]['snippet']['title']))
    return video_list

from time import sleep
for search_query in queries:
    try:
        videos.extend(getVideos(query=search_query, pages=20))
        sleep(3)
    except:
        pass
videos = list(set(videos))

print(len(videos))
for it in videos:
    print(it)
```
