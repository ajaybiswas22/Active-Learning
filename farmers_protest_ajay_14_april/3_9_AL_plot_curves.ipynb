```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [15, 8]
plt.rcParams["figure.autolayout"] = True

df = pd.read_csv('results/02_05_22_80_20_split/results_71_SVM.csv')
df['Seed Size'] = df['Seed Size'].div(100).round(2)

df[['Seed Size','Accuracy','Precision_W','Recall_W','F1score_W']].multiply(100).plot(x='Seed Size')
plt.legend(['Accuracy','Precision','Recall','F1 score'])
plt.ylabel('Percentage')
plt.savefig('outputs/figures/' + 'SVM_20_5_666',dpi=300, bbox_inches='tight')
plt.show()
```
